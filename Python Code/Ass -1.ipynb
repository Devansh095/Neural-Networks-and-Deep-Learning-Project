The Dataset is Highly Unbalanced 

we have to balanced the Dataset, otherwise ML model unable to predict the Fraud Transaction , for that we will use Under Sampling Method

```python
# separating the data for analysis

# 0 >>>> Legitmate Transactions 
# 1 >>>> Fraud Transactions

legit = data[data.Class == 0]
fraud = data[data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

Handling the imbalance Data using SMOTE (Synthetic Minority Oversampling Technique)

we will create New DataSet, which contains similar Distribution of the Legitmate Transactions and Fraud Transaction 

```python
# n = 473 because Fraud Transactions are 473 , so we will take 473 legitmate Transactions

legitmate_sample = legit.sample(n=473)
```

Merging Two DataSet

```python
new_dataset = pd.concat([legitmate_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

SMOTE (Synthetic Minority Oversampling Technique)

```python
from imblearn.over_sampling import SMOTE

# Handle imbalanced data using SMOTE
X = new_dataset.drop('Class', axis=1)
y = new_dataset['Class']
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python

# Count the number of occurrences in each class
class_counts = pd.Series(y_res).value_counts()
print(class_counts)
```

```python
# Plot the class distribution after SMOTE
plt.figure(figsize=(6, 4))
class_counts.plot(kind='bar', color=['green', 'red'])
plt.xticks(ticks=[0, 1], labels=['Legit', 'Fraud'], rotation=0)
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Class Distribution After Applying SMOTE')
plt.show()
```

```python
print(X)
```

```python
print(y)
```

Split the Data into Training and Tesing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training 

Logistic Regression 

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, y_train)
```

Accuracy on Training Data

```python
# Checking accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy Score on Training data : ', training_data_accuracy)
```

Accuracy On Test Data

```python
# Checking accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

Saving a Model to Local

```python
import joblib

# Save the trained model
joblib.dump(model, 'logistic_regression_model.pkl')
```
